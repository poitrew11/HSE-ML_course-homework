```python
print(np.sum(data.isna()))
```

```python
data['18_group'].fillna(0)
data['is_mi'].fillna(0)
data['is_first_time'].fillna('False')
data['17_group'].fillna(0)
data['blended'].fillna(0)
data['is_ml_student'].fillna(0)
```

```python
datata = np.sum(data.isna())
print(datata)
```

```python
answer_distribution = data['is_first_time'].value_counts(normalize = True) * 100
print(answer_distribution)
```

```python
unique_courses = data['blended'].unique()
print(unique_courses)
```

```python
course_counts = data['blended'].value_counts()
print(course_counts.idxmax)
```

```python
average_ratings = data.groupby('blended')['rating'].mean()
print(average_ratings.idxmax())
```

```python
course_columns = ['fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3', 'blended']

course_preferences = data.groupby(course_columns).size().reset_index(name='count')

duplicate_preferences = course_preferences[course_preferences['count'] > 1]
if not duplicate_preferences.empty:
    print("Наборы курсов, выбранные более чем одним студентом:")
    print(duplicate_preferences)
else:
    print("Нет студентов с абсолютно одинаковыми предпочтениями по всем курсам.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = 8, 5
plt.rcParams['font.size'] = 12
sns.set_style('darkgrid')
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
data['day'] = data['timestamp'].dt.date

average_ratings_by_days = data.groupby('day')['rating'].mean().reset_index()

plt.figure(figsize = (10,8))
plt.bar(average_ratings_by_days['day'], average_ratings_by_days['rating'])
plt.title("namer")
plt.legend()
plt.grid()
plt.show()
```
